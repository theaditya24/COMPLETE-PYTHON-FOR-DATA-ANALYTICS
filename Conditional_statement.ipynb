### Q. Validate the quality of email and correctness of email by checking the following conditions :-
###    1. Must not be empty.
###    2. Must contain '.' and '@'
###    3. Must have exaclty one '@'
###    4. Must end with '.com' , '.org' or '.net'
###    5. Must not be longer than 254 characters 
###    6. Must start and end with letter or digit.

```python
email = input("Enter your email : ")
email = email.strip()
if email == "" :
    print("Your email is empty")
elif '.' not in email and '@' not in email:
    print("Your Email is not in the format of aaa@a.com/org/net")
elif email.count('@') != 1:
    print("Your email has more '@' symbol")
# elif email.endswith('.com') or email.endswith('.org') or email.endswith('.net'):
elif not email.endswith(('.com','.org','.net')):
    print("Your email should end with .com/.net/.org")
elif len(email) > 254:
    print('Your email should not exceeds 254 character')
elif not ( email[0].isalnum() and email[-1].isalnum() ):
    print('Your email should only start or end with alphabet and numeric characters only .')
else:
    print('Your email is valid.')
```

### VALIDATE THE QUALITY OF PASSWORDS . PASSWORD SHOULD FOLLOW THE FOLLOWING CONDITIONS:
### 1. MUST NOT BE EMPTY
### 2. MUST BE ATLEAST 8 CHARACTERS LONG
### 3. MUST INCLUDE ATLEAST 1 UPPERCASE AND ATLEAST 1 LOWERCASE CHARACTER
### 4. MUST NOT BE SAME AS THE EMAIL
### 5. MUST NOT CONATAIN ANY SPACES
### 6. MUST START AND END WITH LETTER OR DIGIT

```python
email = input("Enter your email (in format aaa@a.com) : ")
email = email.strip()
password = input("Enter your password : ")
def pass_condition():
    print('''❌ Password should follow these conditions :-
             1. MUST NOT BE EMPTY
             2. MUST BE ATLEAST 8 CHARACTERS LONG
             3. MUST INCLUDE ATLEAST 1 UPPERCASE AND ATLEAST 1 LOWERCASE CHARACTER
             4. MUST NOT BE SAME AS THE EMAIL
             5. MUST NOT CONATAIN ANY SPACES
             6. MUST START AND END WITH LETTER OR DIGIT''')
if password == "":
    pass_condition()
elif len(password) < 8:
    pass_condition()
elif not any(c.isupper() for c in password) or not any(c.islower() for c in password):
    pass_condition()
elif password == email:
    pass_condition()
elif ' ' in password:
    pass_condition()
elif not (password[0].isalnum() and password[-1].isalnum()):
    pass_condition()
else:
    print("Your password is valid")
```

### MATCH CASE IN PYTHON

```python
country = input("enter your country : ").capitalize()
match country:
    case "India":
        print("IN")
    case "United States Of America" | "Usa":
        print("US")
```
