```python
# WAP to count 'A' and 'a' separately from the new.txt
readNew = open("new.txt","r")
readContent = readNew.read()
count_A = 0
count_a = 0
for i in readContent:
    if i == 'a':
        count_a +=1
    elif i == 'A':
        count_A +=1
print("Total number of 'A' is :",count_A)
print("Total number of 'a' is:",count_a)
readNew.close()
```

```python
# WAP to count total uppercase and lowercase characters separately
readNew = open("new.txt","r")
readContent = readNew.read()
total_upper = 0
total_lower = 0
for i in readContent:
    if i.isupper():
        total_upper +=1
    elif i.islower():
        total_lower += 1
print(readContent)
print("total upper char :",total_upper)
print("total lower char :",total_lower)
readNew.close()
```

```python
# wap to count total lowercase and uppercase vowels separately
readNew = open("new.txt","r")
readContent = readNew.read()
total_upper_vowel = 0
total_lower_vowel = 0
for i in readContent:
    if i in 'AEIOU':
        total_upper_vowel += 1
    elif i in 'aeiou':
        total_lower_vowel += 1
print(readContent)
print("Total upper case vowel are :",total_upper_vowel)
print("Total lower case vowel are :",total_lower_vowel)
readNew.close()
```

```python
# print those line which contain atleast 4 words in which there is atleast 4 characters in each word
file = open("new.txt","r")
all_data = file.read()
lines = all_data.split("\n")
for line in lines:
    words = line.split()
    if len(words) >=4:
        count = 0
        for word in words:
            if len(word) >=4:
               count += 1
        if count >=4:
            print(line)
file.close()
```

# 1. What is a CSV File?

 - CSV = Comma Separated Values
 - Stores data in tabular form
 - Each line = one record
 - Values separated by commas 

### 2. CSV Module in Python

```python
import csv
```

### 3. Writing CSV File (List Based)

### 3.1 Write Rows Using List

```python
with open("data.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Name","Age","City"])
    writer.writerow(["Rahul",20,"Delhi"])
```

### 3.2 Write Multiple Rows

```python
rows = [
    ["Amit",21,"Mumbai"],
    ["Neha",19,"Pune"]
]

with open("data.csv", "a", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(rows)
```

### 4. Writing CSV File (Dictionary Based)

### 4.1 Write Using DictWriter

```python
with open("data.csv", "w", newline="") as file:
    fieldnames = ["Name","Age","City"]
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerow({"Name":"Rahul","Age":20,"City":"Delhi"})
```

### 4.2 Write Multiple Dictionary Rows

```python
rows = [
    {"Name":"Amit","Age":21,"City":"Mumbai"},
    {"Name":"Neha","Age":19,"City":"Pune"}
]

with open("data.csv", "a", newline="") as file:
    writer = csv.DictWriter(file, fieldnames=["Name","Age","City"])
    writer.writerows(rows)
```

### 5. Reading CSV File (List Based)

### 5.1 Read Entire CSV as List

```python
with open("data.csv", "r") as file:
    reader = csv.reader(file)
    data = list(reader)
    print(data)
```

### 5.2 Read CSV Row by Row

```python
with open("data.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

### 5.3 Skip Header Row

```python
with open("data.csv", "r") as file:
    reader = csv.reader(file)
    next(reader)
    for row in reader:
        print(row)
```

### 6. Reading CSV File (Dictionary Based)
### 6.1 Read CSV as Dictionary

```python
with open("data.csv", "r") as file:
    reader = csv.DictReader(file)
    for row in reader:
        print(row)
```
