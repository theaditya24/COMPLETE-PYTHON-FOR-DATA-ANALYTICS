```python
l = {"maths":95,
     "sst":90}
l.update({"maths":96, "english":95})  # update() will update the value of existing key-value pair and if the the key is not there in dict then it will add
print(l)
```

```python
print(l.keys())  # by using this you can find all the key present in the dictionary
```

```python
print(l.values()) # by using this you can find all the values present in the dictionary
```

```python
print(l.get("sst")) # it will get the value of specific key from the dictionary. 
# If the entered key is not present in the dictionary then it will return None
# HERE IS THE CATCH - IF WE USE DICT_NAME[KEY_NAME] TO FIND THE VALUE OF THAT SPECIFIC KEY AND IF THE KEY IS NOT THERE THEN IT WILL GIVEN AN 'ERROR'.
# WHEREAS IF WE USE DICT_NAME.GET(KEY_NAME) AND IF THAT KEY NAME IS NOT PRESENT IN THE DICTIONARY THEN IT WILL GIVE US THE 'NONE'.
```

```python
# merging two dictionaries
d1 = {"a": 10, "b": 20}
d2 = {"c": 30, "d": 40}
d1.update(d2)
print(d1)

# Method 2: Using ** Unpacking (Very Popular) - Best for making a new dictionary without changing old ones.
# merged = {**d1, **d2}
# print(merged)

# Important Case: If Both Have Same Key - In this case the value from d4 replaces the value from d3.
d3 = {"a": 10, "b": 20}
d4 = {"b": 100, "c": 30}
merged = {**d3, **d4}
print(merged)
```

```python
# convert two list into a dict
keys = ["name", "age", "city"]
values = ["Aditya", 22, "Delhi"]

# METHOD -1 - USNG ZIP() IT BASCIALLY MATCHES THE 0TH INDEX OF ONE LIST TO ANOTHER LIST AND THUS FORM A DICT
result = dict(zip(keys, values))
print(result)

# METHOD -2 BY USING FOR LOOPS AND DOING IT MANUALLY
# result = {}
# for i in range(len(keys)):
#     result[keys[i]] = values[i]
# print(result)
```

```python
# Word Frequency Counter
sentence = "python is easy and python is powerful"
lstOfSent = sentence.split()
freq = {}
for word in lstOfSent:
    if word in freq:
        freq[word] +=1
    else:
        freq[word] = 1
print(freq)
```
