### FUNCTION IN PYTHON

### 1. Defining a Basic Function

```python
def function_name():
    # Function body: code to execute
    print("Hello, this is inside the function!")

def greet():
    print("Hello, World!")

# Calling the function
greet()  # Output: Hello, World!
```

### 2. Parameters and Arguments

```python
# POSITIONAL AND KEYWORD PARAMETER
def add_numbers(a, b):  # a and b are parameters
    result = a + b
    print(f"The sum is: {result}")

# Calling with arguments
add_numbers(5, 3)      # Positional: Output: The sum is: 8
add_numbers(b=3, a=5)  # Keyword: Same output
```

```python
# Default Parameters Example:
def greet_user(name="Guest"):  # Default value for name
    print(f"Hello, {name}!")

greet_user()           # Output: Hello, Guest!
greet_user("Alice")    # Output: Hello, Alice!
```

```python
# Variable-Length Arguments Example:
def sum_all(*args):  # *args collects extras into a tuple
    total = 0
    for num in args:
        total += num
    print(f"Sum: {total}")

sum_all(1, 2, 3, 4)  # Output: Sum: 10

def print_info(**kwargs):  # **kwargs collects into a dict
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Bob", age=25)  # Output: name: Bob \n age: 25
```

### 3. Return Values

```python
def multiply(x, y):
    return x * y  # Returns the product

result = multiply(4, 5)
print(result)  # Output: 20

# Multiple returns (only one executes)
def check_even(n):
    if n % 2 == 0:
        return "Even"
    return "Odd"  # Default return

print(check_even(4))  # Output: Even
```

### 4. Scope and Lifetime

```python
x = 10  # Global variable

def modify():
    x = 20  # Local variable (shadows global)
    print(f"Inside: {x}")  # Output: 20

modify()
print(f"Outside: {x}")  # Output: 10 (global unchanged)

def modify_global():
    global x
    x = 30  # Now modifies global

modify_global()
print(f"After global: {x}")  # Output: 30
```

### 5. Lambda Functions (Anonymous Functions)

```python
square = lambda x: x ** 2
print(square(3))  # Output: 9

# With map()
numbers = [1, 2, 3]
squares = list(map(lambda n: n**2, numbers))
print(squares)  # Output: [1, 4, 9]
```
