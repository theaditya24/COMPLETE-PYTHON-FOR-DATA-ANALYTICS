### LIST IN PYTHON

A list is a collection of items (like numbers, strings, or even other lists) that is:

Ordered → items have a fixed position (index)

Mutable → you can change, add, or remove items

Defined using square brackets [ ]

```python
# Creating a list
fruits = ["apple", "banana", "cherry"]
print(fruits)
```

```python
# Another method where we can store data in a list. List() -> it is a Built-in function . 
# It basically convert all iterable sequence of characters into list.
letters = 'Aditya Raj'
print(list(letters))
```

```python
# Accessing a list
fruits = ["apple", "banana", "cherry"]
print(fruits[0])   # apple
print(fruits[1])   # banana
print(fruits[-1])  # cherry (negative index = from the end)
```

```python
fruits = ["apple", "banana", "cherry"]
print(fruits[::1]) # output = ['apple', 'banana', 'cherry']
print(fruits[:1])  # output = ['apple']
print(fruits[::-1])  # output = ['cherry', 'banana', 'apple'] -> it basically returns a new list not modifies the original one
fruits.reverse()  # it will also reverse the list but it will change/modify the original one.
print(fruits)
```

In python, it stores list in the form object which have two parts first one is its type (i.e. list) and second one is its value.
Our variable which has been assigned to a list it points to that object .
If we take the above example :- fruits will point to each object that has been created for this list in memory. For each data in list there will be separate object in memory and the variablw will be pointing to each one. 

```python
# modifying list
fruits[1] = "mango"
print(fruits)      # ['apple', 'mango', 'cherry']
```

```python
# Adding items
fruits.append(1)
print(fruits)
```

```python
# insert at a specific position
fruits.insert(1, "grapes")
print(fruits)
```

```python
# looping through list
for item in fruits:
    print(item)
```

### PACKING AND UNPACKING IN LIST

```python
data = ['aditya', '22', 'Data Analyst','delhi']
Name, Age, Role, Location = data  # This will here automatically bring the data from data List and put it in order wise
print(Name)
print(Age)
print(Role)
```

```python
# Extended Unpacking with *
# Unpack first/last explicitly, rest into a list
l = [1, 2, 3, 4, 5]
head, *middle, tail = l
print(head)    # 1
print(middle)  # [2, 3, 4]
print(tail)    # 5
```

```python
# Packing
packed_list = [10, 20, 30, 'hello']

# Unpacking
x, y, z, greeting = packed_list
print(f"Unpacked: x={x}, y={y}, z={z}, greeting={greeting}")

# Extended unpacking
first, *rest = packed_list
print(f"First: {first}, Rest: {rest}")

# Packing back (e.g., unpack and repack)
repacked = [x, y, z]
print(f"Repacked: {repacked}")
```
