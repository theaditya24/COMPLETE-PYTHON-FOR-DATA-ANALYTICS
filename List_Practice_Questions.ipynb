# Challenge 1 — Odd/Even Counter

```python
exList = [1,2,3,4,5]
evenCount = 0
oddCount = 0
for number in exList:
    if number % 2 == 0:
        evenCount += 1
    else:
        oddCount += 1
print(f'Total Even Numbers in the given list is = " {evenCount} " and Total Odd numbers in the given list is = " {oddCount} "')
```

# Challenge 2 — Sum and Average

```python
exList = [1,2,3,4,5]
sumofList = 0
# avgofList = 0
for number in exList:
    sumofList += number
avgofList = sumofList/len(exList)
print(f'Sum = {sumofList} and average = {avgofList}')
```

# 🔢 Challenge 3 — Find Maximum and Minimum

```python
exList = [1,2,3,10,5]
maxItem = exList[0]
minItem = exList[0]
for number in exList:
    if maxItem < number:
        maxItem = number
    elif minItem > number:
        minItem = number
print(maxItem)
print(minItem)
```

# 🔁 Challenge 4 — Reverse a List

```python
exList = [1, 2, 3, 10, 5]

for i in range(len(exList) // 2):  # i=0,1 (half of 5)
    exList[i], exList[len(exList) - 1 - i] = exList[len(exList) - 1 - i], exList[i]

print(exList)  # Output: [5, 10, 3, 2, 1]
```

# 📦 Challenge 5 — Count Positive, Negative, and Zero

```python
l = [-2 ,0, 5, -8, 0, 7]
pos = 0
neg = 0
zero = 0
for i in l:
    if i > 0:
        pos += 1
    elif i < 0:
        neg += 1
    else:
        zero += 1
print(f'Positive number = {pos} \nNegative Number = {neg} \nZero = {zero}')
```

# 🧠 Challenge 8 — Palindrome Checker

```python
def is_palindrome(s):
    cleaned = ''.join(char.lower() for char in s if char.isalnum())  # this will basically first check whether all the characters of string contain,
    # only alphabet and numeric characters only. And if they do then it will convert it to lower case and in last it will join them all.
    # and if any character is not alphanumeric it will ignore that character and will not join that in the final string. 

    left,right = 0 , len(cleaned)-1
    while left < right:
        if cleaned[left] != cleaned[right]:
            return False
        left += 1
        right -= 1
        return True

user_input = input("Enter a string to check to whether it is palindrome or not : ")
if is_palindrome(user_input):
    print("It's a palindrome!")
else:
    print("It's not a palindrome.")
```
