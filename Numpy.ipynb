# RANDOM MODULE IN PYTHON

### 1️⃣ random.random() - Returns a random floating number between 0 and 1.

```python
import random
```

```python
print(random.random())
```

### 2️⃣ random.randint() - Returns a random integer between two numbers (inclusive).

```python
print(random.randint(1,10)) 
# Range - a ≤ value ≤ b
```

### 3️⃣ random.randrange() - Returns a random number from a range.

```python
print(random.randrange(1,10))
# Range - start ≤ value < stop
# Syntax - random.randrange(start, stop, step)
```

### 4️⃣ random.choice() - Selects one random element from a sequence.

```python
# Syntax - random.choice(sequence)
colors = ["red","blue","green"]

print(random.choice(colors))
```

### 5️⃣ random.choices() - Returns multiple random elements with replacement.

```python
data = [1,2,3,4]

print(random.choices(data, k=2))

# with replacement means DUPLICATES ARE ALLOWED
```

### 6️⃣ random.sample() - Returns multiple random elements without replacement.

```python
data = [1,2,3,4,5]

print(random.sample(data,3))

# without replacement means DUPLICATES ARE NOT ALLOWED
```

### 7️⃣ random.shuffle() - Randomly shuffles elements of a list.

```python
data = [1,2,3,4]

random.shuffle(data)

print(data)

# it changes the original list
# ⚠ Works only with lists
```

### 8️⃣ random.uniform() - Returns a random floating number between two values.

```python
print(random.uniform(5,10))

# Range - 5 ≤ value ≤ 10
```
