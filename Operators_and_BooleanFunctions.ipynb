BOOLEAN VALUES :- TRUE FALSE

BOOLEAN FUNCTIONS :- 1. bool(value) = it is a built-in function = True : if value is non-zero or non-empty ,
                                                                  False : if value is empty or zero
                     2. all(iterable) = Returns True if all elements in the iterable are truthy; otherwise False
                                Empty iterable: True (vacuously true—no elements to falsify).
                                Useful for validating entire datasets (e.g., all values > 0)., 
                     3. any(iterable) = Returns True if any element in the iterable is truthy; otherwise False
                                Empty iterable: False (no elements to make it true).
                                Great for quick checks (e.g., any missing data?), 
                     4. isinstance(obj, class_or_tuple) = Returns True if obj is an instance of class_or_tuple (or subclass); else False.

```python
print(type(True)) # here type() is a function which tells us the type of value which has been provided inside that function
```

```python
print(bool(123)) # output = True
print(bool())    # output = False
print(bool(0))   # output = False
```

```python
numbers = [1, 2, 3]  # All truthy
print(all(numbers))  # True

zeros = [1, 0, 3]    # Has a falsy (0)
print(all(zeros))    # False

empty = []           # Empty is True
print(all(empty))    # True

email = ""
phone = "929292"
password = ""

print(all([email,phone,password]))   # Here it will check if all the values of list have there own value or not. 
                                     # If one of them is empty or zero then it will return False.
```

```python
numbers = [0, 0, 5]  # Has a truthy (5)
print(any(numbers))  # True

all_zeros = [0, 0]   # All falsy
print(any(all_zeros)) # False

empty = []           # Empty is False
print(any(empty))    # False

email = ""
phone = "929292"
password = ""

print(any([email,phone,password]))   # Here it will check if any the values of list have there own value or not. 
                                     # If one of them is non-empty or non-zero then it will return true.
```

```python
value = 42
print(isinstance(value, int))      # True

text = "hello"
print(isinstance(text, (int, str))) # True (matches str in tuple)

wrong = [1, 2]
print(isinstance(wrong, float))    # False
```

Comparison operator in python -: Comparison operators in Python are used to compare two values and return a boolean result (True or False).

1. == (Equal)
Description: Returns True if the values are equal.

2. != (Not Equal)
Description: Returns True if the values are not equal.

3. < (Less Than)
Description: Returns True if the left value is less than the right.

4. '>' (Greater Than)
Description: Returns True if the left value is greater than the right.

5. <= (Less Than or Equal)
Description: Returns True if the left value is less than or equal to the right.

6. '>=' (Greater Than or Equal)
Description: Returns True if the left value is greater than or equal to the right.

```python
print(5 == 5)      # True (same number)
print("hello" == "hello")  # True (same string)
print(3 == 4)      # False

print(10 != 7)     # True (different)
print([1, 2] != [1, 2])  # False (same list content)
print("cat" != "dog")  # True

print(3 < 5)       # True
print("apple" < "banana")  # True ('a' before 'b')
print(10 < 2)      # False

print(8 > 4)       # True
print("zebra" > "ant")  # True ('z' after 'a')
print(1 > 10)      # False

print(7 <= 7)      # True (equal)
print(2 <= 5)      # True (less)
print(9 <= 3)      # False

print(6 >= 6)      # True (equal)
print(10 >= 3)     # True (greater)
print(4 >= 8)      # False
```

### Logical Operator

### 1. and - Returns True if both operands are truthy; else False.
### 2. or - Returns True if at least one operand is truthy; else False.
### 3. not - Inverts the truth value of its operand (True → False, False → True)

```python
a = True
b = False
print(a and b)  # False (short-circuits after False)

x = 5 > 3  # True
y = 10 < 2 # False
print(x and y)  # False

a = True
b = False
print(a or b)  # True (short-circuits after True)

x = 5 > 3  # True
y = 10 < 2 # False
print(x or y)  # True

value = True
print(not value)  # False

condition = 5 < 3  # False
print(not condition)  # True
```
