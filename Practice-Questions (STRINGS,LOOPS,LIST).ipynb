# 1. Text Analyzer Tool (BEST for learning)

### ✔ Count words
### ✔ Count characters
### ✔ Count sentences
### ✔ Most frequent word
### ✔ Remove punctuation

### Features:

### User enters a paragraph

### Your code analyzes it using loops + string functions + lists

```python
para = input("Enter any paragraph / word : ").strip()
# print(para)
# 1. Remove punctuation
punctuations = ".,!?;:-_()[]{}\"'"
clean_text = ""
for char in para:
    if char not in punctuations:
        clean_text += char
# print(clean_text)

# 2. Count Words
para_length = len(clean_text.split(" "))
print(f"Total number of words in the Para you entered are : {para_length}")

# 3. Count characters
char_count = 0
for char in clean_text:
    if char != " ":
        char_count += 1
print(f"Total number of Characters in the Para you entered are : {char_count}")

# 4. Count Sentences
sentence_count = 0
for char in para:
    if char in ".!?":
        sentence_count += 1
print(f"Total number of Sentences in the Para you entered are : {sentence_count}")

# 5. Most frequent word
freq = {}
for w in clean_text:
    w = w.lower()
    if w in freq:
        freq[w] += 1
    else:
        freq[w] = 1

# find word with highest count
most_frequent = max(freq, key=freq.get)
print("Most Frequent Word:", most_frequent)
```

# 2. To-Do List App (Console)

### ✔ Add tasks
### ✔ Remove tasks
### ✔ View tasks
### ✔ Mark complete

```python
print("========================")
print("\tTO-DO LIST")
print("========================\n")

print('''1. Add Task
2. View Tasks
3. Remove Task
4. Exit\n''')

task = []

while True:
    task_number = int(input("Choose an option from the above : "))
    
    if task_number == 1:
        new_task = input("Enter a New Task you want to Add in To-Do list : ")
        task.append(new_task)
        print("\n Task Added Successfully!!")
        
    elif task_number == 2:
        if len(task) == 0:
            print("No tasks found!\n")
        else:
            print("\nYour Tasks are:")
            for i in range(len(task)):
                print(f"{i+1}. {task[i]}")
            print()
            
    elif task_number == 3:
        del_task = int(input("Enter the task number which you want to delete :"))
        if len(task) == 0:
            print("No tasks available to delete!\n")
        else:
            del_task = int(input("Enter the task number which you want to delete: "))
            if 1 <= del_task <= len(task):
                task.pop(del_task - 1)
                print("Task Deleted Successfully!\n")
            else:
                print("Invalid task number!\n")

    # 4. Exit
    elif task_number == 4:
        print("Exiting program... Thank you!")
        break

    else:
        print("Invalid option! Please enter 1–4.\n")
```

# 3. BASIC CALCULATOR

```python
print("========================")
print("\tTO-DO LIST")
print("========================\n")

num = int(input("Enter how many numbers you want to enter into Calculator : "))
num_list = []

for i in range(1,num+1):
    number = int(input(f"Enter the number {i} : "))
    num_list.append(number)

print('''\n1. Addition
2. Subtarction
3. Multiplication
4. Divison
5. Exit\n''')

while True:
    task_number = int(input("Choose an option from the above : "))

    if task_number == 1:
        # sum_list = 0
        # for i in range(len(num_list)):
        #     sum_list += num_list[i]
        print("Sum of Numbers you have entered is : ",sum(num_list))

    # 2. Subtraction
    elif task_number == 2:
        result = num_list[0]              # start from first number
        for i in range(1, len(num_list)): # subtract the rest
            result -= num_list[i]
        print("Result of Subtraction is :", result)

    # 3. Multiplication
    elif task_number == 3:
        result = 1
        for n in num_list:
            result *= n
        print("Result of Multiplication is :", result)

    # 4. Division
    elif task_number == 4:
        result = num_list[0]
        division_possible = True
        for i in range(1, len(num_list)):
            if num_list[i] == 0:
                print("Error: Cannot divide by zero! One of the numbers is 0.")
                division_possible = False
                break
            result /= num_list[i]
        if division_possible:
            print("Result of Division is :", result)

     # 4. Exit
    elif task_number == 5:
        print("Exiting program... Thank you!")
        break

    else:
        print("Invalid option! Please enter 1–5.\n")
```

# 4. Count the Vowels

```python
sentence = input("Enter a Sentence and i will count the number of times vowels are coming : ").lower()
count = 0
vowel = []

for char in sentence:
    if char in 'aeiou':
        count += 1
        vowel.append(char)
print(f"Vowels found : {vowel} | Count: {count} ")
```

# 6. Find the Longest Word in a Sentence

```python
# sentence = input("Enter a sentence: ")

# words = sentence.split()

# longest_word = [words[0]]
# for word in words:
#     if len(word) > len(longest_word[0]):
#         longest_word.pop()
#         longest_word.append(word)
# print(f"Longest Word : {longest_word[0]} ( Length : {len(longest_word[0])} )")

sentence = input("Enter a sentence: ")

words = sentence.split()

# Handle empty input case
if not words:
    print("No words found.")
else:
    # Step 1: Find the maximum length
    max_len = 0
    for word in words:
        if len(word) > max_len:
            max_len = len(word)

    # Step 2: Collect all words having max length
    longest_words = []
    for word in words:
        if len(word) == max_len:
            longest_words.append(word)

    # Step 3: Print results
    print(f"Longest word: {', '.join(longest_words)}")
    print(f"Length: {max_len}")
```
