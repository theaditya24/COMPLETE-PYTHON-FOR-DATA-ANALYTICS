### ✅ What is a Set?

A set is a collection of items that is:

Unordered

Unindexed

No duplicate values allowed

Stores only unique elements

```python
nums = {1, 2, 3, 3, 4}
print(nums)
# 📌 Duplicate 3 is automatically removed.
```

### ✅ Why Use Sets?

Sets are mainly used for:

Removing duplicates

Membership checking (in)

Mathematical operations (union, intersection)

```python
# creating an empty set
s = set()
print(s)
```

### ✅ Set Methods with Easy Examples

```python
# 1. add() → Add one element
s = {1, 2, 3}
s.add(4)
print(s)
```

```python
# 2. update() → Add multiple elements
s = {1, 2}
s.update([3, 4, 5])

print(s)
```

```python
# 3. remove() → Remove element (Error if not found)
s = {1, 2, 3}
s.remove(2)

print(s)
```

```python
# 4. discard() → Remove element (No error)
s = {1, 2, 3}
s.discard(10)

print(s)
```

```python
# 5. pop() → Removes random element
s = {10, 20, 30}
s.pop()

print(s)
```

```python
# 6. clear() → Remove all elements
s = {1, 2, 3}
s.clear()

print(s)
```

### ✅ Set Operations (Very Important)

```python
# 7. Union (| or union()) - Combine all unique elements.

a = {1, 2, 3}
b = {3, 4, 5}

print(a.union(b))
```

```python
# 8. Intersection (& or intersection()) - Common elements only.

a = {1, 2, 3}
b = {2, 3, 4}

print(a.intersection(b))
```

```python
# 9. Difference (- or difference()) - Elements present in first set but not in second.

a = {1, 2, 3}
b = {2, 4}

print(a.difference(b))
```

```python
# 10. Symmetric Difference - Elements that are NOT common.
a = {1, 2, 3}
b = {2, 3, 4}

print(a.symmetric_difference(b))
```

```python
# ✅ Membership Check
s = {10, 20, 30}

print(20 in s)
print(50 in s)
```

```python
nums = [1, 2, 2, 3, 4, 4, 5]

unique_nums = set(nums)

print(unique_nums)
```
