```python
# count total lower case vowels
string = input("enter any word : ")
count = 0
for char in string:
    if char in 'aeiou' and char.islower():
        count += 1
print(count)
```

```python
# count total lower case consonants
string = input("enter any word : ")
count = 0
for char in string:
    if char not in 'aeiou' and char.islower():
        count += 1
print(count)
```

```python
string = input("enter any word : ")
Vowel_count = 0
Cons_Count = 0
for char in string:
    if char in 'aeiou':
        Vowel_count += 1
    elif char not in 'aeiou' and char.islower():
        Cons_Count += 1
print("Vowel Count : ",Vowel_count)
print("consonant count : ",Cons_Count)
```

# STRING FUNCTIONS

### 1. Basic String Info Functions

```python
# len() — length of string
s = "Data Science"
print(len(s))   # 12

# 2. type() — check data type
type("hello")   # <class 'str'>
```

### 2. Case Conversion Functions

```python
# 3. lower()
"DATA".lower()   # 'data'

# 4. upper()
"data".upper()   # 'DATA'

# 5. title()
"machine learning".title()   # 'Machine Learning'

# 6. capitalize()
"python".capitalize()   # 'Python'
```

### 3. Search & Check Functions

```python
# 7. startswith()
"name.csv".startswith("name")   # True

# 8. endswith()
"report.pdf".endswith(".pdf")   # True

# 9. find() - Returns index, or -1 if not found.
"data analytics".find("ana")   # 5

# 10. count()
"banana".count("a")   # 3

# 11. isalnum(), isalpha(), isdigit()
"123".isdigit()   # True
"name123".isalnum()   # True
"abc".isalpha()   # True
```

### 4. Replace & Remove

```python
# 12. replace() - it does not change the original string it basically duplicate the original string and create a copy of that string and then do all the functions on that string
                #  so hence it returns that string on which the functions has been performed.
"text data".replace(" ", "_")   # 'text_data'
string = "Doto Analyst"   # let's say we have found a string where Data is written as Doto and we want to fix it
# update = string.replace('Doto','Data')
update = string.replace('o','a',2)  # here 2 is the count of 'o' that has to be replaced from the start of the string
print(update)

# 13. strip() - also lstrip() – left spaces and rstrip() – right spaces
"  hello  ".strip()   # 'hello'
```

### 5. Splitting & Joining (Most Important in Data Science)

```python
# 14. split() - Break string into list.
    # string.split(separator, maxsplit)
    # separator → where to cut the string
    # maxsplit → how many times to cut (optional)
    # If you don’t give a separator, Python splits on spaces by default. And if you don't give any value of maxsplit it will run through whole string
"python,data,science".split(",")
# ['python', 'data', 'science']
new = "Aditya Raj".split()  # by default it will cut by spaces
print(new)
# If separator is not found
"hello".split(",")  # ['hello']   # No split happens
# using maxsplit argument
text = "a b c d"
text.split(" ", 2) # ['a', 'b', 'c d']

# 5. splitlines() - Useful for reading logs or text files.
"line1\nline2".splitlines()

# 16. join() - takes a list, and puts a symbol between every element.
    # Syntax - separator.join(iterable) 
    # separator → the string you want between elements
    # iterable → list, tuple, etc.

",".join(["a","b","c"])    # 'a,b,c'
# It means:
# join the list ["a", "b", "c"]
# put , between each element

" ".join(["Data", "Science"]) # 'Data Science'

words = ["Python", "is", "fun"]
" ".join(words)  # 'Python is fun'

"".join(["A", "B", "C"]) # 'ABC'

"-".join("DATA") # 'D-A-T-A'
```

### 6. String Formatting (Very Important)

```python
# WHAT IS STRING FORMATTING?
# String formatting means:
# Putting variables (values) inside a string in a clean and readable way.
# EX- name = "Aditya"
    # age = 22
# We want to print: My name is Aditya and I am 22 years old.
# SO FOR THIS WE HAVE TOTAL 3 METHOD IN PYTHON : i) format()  (ii) %   (iii) f-string - modern and very important

# 17. format()
"Name: {}, Age: {}".format("Aditya", 22)

# 18. f-strings (Recommended)
name = "Aditya"
print(f"Hello {name}")
```
