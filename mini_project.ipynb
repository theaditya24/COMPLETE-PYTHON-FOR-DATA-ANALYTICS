```python
from pathlib import Path
import os
import shutil
```

```python
def create_folder():
    try:
        name = input("Enter the name of folder : ")
        p = Path(name)
        p.mkdir()
        print("folder created successfully")
    except Exceptions as e:
        print(f"An error occurred which is {e}")
```

```python
def read_folder():
    p = Path("")
    items = list(p.rglob("*"))
    for item,value in enumerate(items):
        print(f"{item + 1} : {value}")
```

```python
def update_folder():
    try:
        read_folder()
        old_name = input("From the above list of items(folder) Which one do you want to update ? ")
        p = Path(old_name)
        if p.exists() and p.is_dir():
            new_name = input("Enter the name of for your new Folder : ")
            new_p = Path(new_name)
            p.rename(new_p)
            print("Folder name updated successfully !!")
        else:
            print("No folder of entered name exists ")
    except Exceptions as e:
        print(f"An error occurred which is {e}")
```

```python
def del_folder():
    try:
        read_folder()
        name = input("Enter the name of folder you want to DELETE from the above list : ")
        p = Path(name)
        if p.exists() and p.is_dir():
            shutil.rmtree(name)
            print("Folder deleted successfully !!")
        else:
            print("No such folder exists")
    except Exception as e:
        print(f"An error occurred {e}")
```

```python
def create_folder():
    try:
        read_folder()
        name = input("Enter the name of file you want to create : ")
        p = Path(name)
        if not p.exists():
            with open(name,"w") as file:
                data = input("enter the data you want to add in the file : ")
                file.write(data)
            print("file created successfully")
        else:
            print("OOPS!! this file already exists")
    except Exception as e:
        print(f"An error occurred : {e}")
```

```python
print("---- SELECT ANY ONE OF THE FOLLOWING ----")
print("1. Create a folder ")
print("2. Read files and folder ")
print("3. Update the folder ")
print("4. Delete the folder ")
print("5. Create a file ")

choice = int(input("enter your choice from the above :- "))

if choice == 1:
    create_folder()
if choice == 2:
    read_folder()
if choice == 3:
    update_folder()
if choice == 4:
    del_folder()
if choice == 5:
    create_folder()
```
